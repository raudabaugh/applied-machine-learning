```python
%matplotlib inline

import skimage as ski
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from skimage import io
from skimage import filters
from skimage import color
from skimage.morphology import disk
from random import randint
import numpy as np
```

```python
img = ski.io.imread("mona-lisa.jpg")
img = color.rgb2gray(img)
```

```python
def coordinate_to_position(x,y, shape):
    columns = shape[0]
    rows = shape[1] 
    return x*rows + y
```

```python
samples = []
labels = []

i = 0
while i < 5000:
    x = randint(0,899)
    y = randint(0,603)
    
    samples.append(img[x][y])
    labels.append([x,y])
    
    i += 1
```

```python
x = labels
y = samples
```

```python
labels
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators=10)
clf = clf.fit(x, y)
```

```python
clf.predict([[0,0]])
```

```python
predictions = []

for i in range(0,900):
    for j in range(0,604):
        predictions.append([i,j])
```

```python
predicted = clf.predict(predictions)
```

```python
predicted_img = predicted.reshape(900,604)
```

```python
plt.imshow(predicted_img, cmap=cm.gray)
```
